```python
from utils.scatter_normal import PlotManager
import matplotlib.pyplot as plt
import numpy as np

plt.rc('font', family='serif')
plt.rc('font', serif='CMU Serif')

x = np.linspace(1, 10, 20)
y = 2 * x + np.random.normal(0, 2, len(x))

fig, axes = plt.subplots(2, 3, figsize=(8, 5))
for ax in axes.flatten():
    pm = PlotManager(ax, fontsize=12, fontsize_s=9, scale='linear', xmin=0, xmax=12)
    pm.ax_setting()
    pm.scatter_plot(x, y, color='k')
    pm.plot_regression_with_ci(x, y, color='k', text_loc=(0.95, 0.35))

axes[0, 0].set_xlabel('Year', fontsize=14); axes[0, 0].set_ylabel('Capacity Loss [%]', fontsize=14)

plt.tight_layout()
plt.show()
```
